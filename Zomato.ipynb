```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#convert ratings
df['rate'].unique()
```

```python
#Replace 'NEW' and '-'  with NaN
df['rate'] = df['rate'].replace('NEW',np.nan).replace('-',np.nan)
#Remove '/5' and convert to float
df['rate'] = df['rate'].str.split('/').str[0].astype(float)
```

```python
df['rate'].head
```

```python
#check datatypes
df.dtypes
```

```python
#Data Analysis and visualization
#Types of restaurants
#Count plot of restaurant types
sns.countplot(data=df,x='listed_in(type)')
plt.title('Count of distribution of restaurant types')
plt.show()
```

```python
#Votes by restaurant type
#Group by 'listed_in(type)' and sum the votes
votes_by_type = df.groupby('listed_in(type)')['votes'].sum().sort_values(ascending=False)

#Plot the result
plt.figure(figsize=(12, 6))
votes_by_type.plot(kind='bar', color='skyblue')

plt.title('Total Votes per Restaurant Type')
plt.xlabel('Restaurant Type')
plt.ylabel('Total Votes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Ratings distribution
#Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(df['rate'], bins=20, color='lightgreen', edgecolor='black')

plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Restaurants')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#Restaurant cost preferance
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='approx_cost(for two people)')

plt.title('Cost Preference for Couples')
plt.xlabel('Approx Cost (for two people)')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#online v/s offline ratings
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='online_order', y='rate')

plt.title('Comparison of Ratings: Online vs Offline Orders')
plt.xlabel('Online Order Available')
plt.ylabel('Rating')
plt.tight_layout()
plt.show()
```

```python
#online orders by restaurant type
#Create a pivot table
pivot_table = pd.pivot_table(
    df, 
    values='name',  # or any column (we're counting rows)
    index='listed_in(type)', 
    columns='online_order', 
    aggfunc='count', 
    fill_value=0
)

# Step 2: Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlGnBu', linewidths=0.5)

plt.title('Number of Restaurants (Online vs Offline) by Type')
plt.xlabel('Online Order Available')
plt.ylabel('Restaurant Type')
plt.tight_layout()
plt.show()
```
